### 1. Merge the customers table with the orders table and give the customer address and the order date for the number of orders being larger than 10.

```python
merge_1 = df_orders.merge(df_customers,on='cust_id',how='inner')
merge_1[merge_1.order_num > 10][['cust_address','order_date']]
```

### 2. Merge the vendors table and the customers table on vend_country and cust_country. Give out the vend_address and the vend_city for those entries where the cust_contact is John Smith.

```python
merge_2 = df_vendors.merge(df_customers, left_on=df_vendors['vend_country'],
                 right_on=df_customers['cust_country'],how='inner')
merge_2[['vend_address','vend_city']][merge_2.cust_contact=='John Smith']
```

### 3. Merge the vendors table and the product table on vend_id and give out vend_name, prod_name and prod_price. Produce a pivot_table with vend_name as index and prod_price as value using the aggregate function np.sum.

```python
merge_3 = (df_vendors.merge(df_products, left_on=df_vendors['vend_id'],
                 right_on=df_products['vend_id'],how='inner'))[['vend_name','prod_name','prod_price']]
merge_3
```

```python
pd.pivot_table(merge_3,index='vend_name',values=['prod_price'],aggfunc=np.sum)
```

### 4. Merge the products table and the ordersitems table on prod_id and give out vend_id, item_price and quantity. From the result, produce a single pivot table with index vend_id and values item_price and quantity using the aggregate functions np.mean, np.std and np.sum.Â¶

```python
merge_4 = (df_products.merge(df_ordersitems, left_on=df_products['prod_id'],
                 right_on=df_ordersitems['prod_id'],how='inner'))[['vend_id','item_price','quantity']]
merge_4
```

```python
merge_4.pivot_table(index='vend_id',values=['item_price','quantity'],aggfunc=[np.mean,np.std,np.sum])
```
