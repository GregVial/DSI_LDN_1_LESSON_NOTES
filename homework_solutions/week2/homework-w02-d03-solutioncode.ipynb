Read in the following dataset collecting information about nutrition habits of 54 individuals.

2. Get dummy variables for the columns of binary categorical type and retain only those columns which are necessary for desribing the data. How many are those?
3. Get dummy variables for the integer valued columns ['BREAKFASTSANDWICHFREQ', 'BREAKFASTSANDWICHQUAN']. Again, retain only those columns which are necessary for describing the data. How many are those? Rename the columns suitably with the command pd.rename.
4. Join the dataframes for all the dummy variables you created with pd.join.

```python
import pandas as pd
import numpy as np

# Import dataset
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/nutrition-studies/raw_anonymized_data.csv')
df.head()
```

```python
# Drop unnecessary ID variable and print the shape of the dataframe
df.drop('ID',inplace=True,axis=1)
df.shape
```

```python
# the number of columns with different datatypes
df.get_dtype_counts()
```

```python
# get the names of the corresponding columns
col_object = []
col_int = []
col_float = []
for col in df.columns:
    if df[col].dtype == int:
        col_int.append(col)
    elif df[col].dtype == float:
        col_float.append(col)
    elif df[col].dtype == object:
        col_object.append(col)
    else:
        print col

print len(col_object), len(col_int), len(col_float)
```

```python
# Get dummies for the binary categorical variables, drop unnecessary columns
df_obj_dummies = (pd.get_dummies(df[col_object]))
df_obj = df_obj_dummies[df_obj_dummies.columns[1::2]]
```

```python
# Get dummies for 'BREAKFASTSANDWICHFREQ' and rename the columns, no columns to drop because not all 
# values within the range occur
print 'Unique values:', sorted(df[col_int[0]].unique())
BSF_dummies = pd.get_dummies(df[col_int[0]])
BSF_names = {name:'BSF_'+str(name) for name in BSF_dummies.columns}
BSF_dummies.rename(columns=BSF_names,inplace=True)
BSF_dummies.head()
```

```python
# Get dummies for 'BREAKFASTSANDWICHQUAN' and rename the columns, there are only two unique values, 
#drop unnecessary column
print 'Unique values:', sorted(df[col_int[1]].unique())
BSQ_dummies = pd.get_dummies(df[col_int[1]])
BSQ_dummies.drop([2],axis=1,inplace=True)
BSQ_names = {name:'BSQ_'+str(name) for name in BSQ_dummies.columns}
BSQ_dummies.rename(columns=BSQ_names,inplace=True)
BSQ_dummies.head()
```

```python
# Join the three dataframes
df_obj = df_obj.join(BSF_dummies)
df_obj = df_obj.join(BSQ_dummies)

print df_obj.shape
df_obj.head()
```
