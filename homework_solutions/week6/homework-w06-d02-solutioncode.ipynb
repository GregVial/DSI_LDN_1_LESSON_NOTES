```python
%%sql 
SELECT 
e."EmployeeID", r."RegionDescription", 
t."TerritoryDescription", o."OrderID", 
od."ProductID", p."UnitPrice" 
FROM employeeterritories e
INNER JOIN territories t ON 
e."TerritoryID" = t."TerritoryID"
INNER JOIN region2 r ON 
t."RegionID" = r."RegionID"
INNER JOIN orders o ON
e."EmployeeID" = o."EmployeeID"
INNER JOIN order_details od ON
o."OrderID" = od."OrderID"
INNER JOIN products p ON 
od."ProductID" = p."ProductID"
ORDER BY 1, 2, 3;
```

#### 3. What is the total price of units sold by the individual employees?

```python
%%sql 
SELECT a1."EmployeeID", SUM(a1."UnitPrice") FROM
(SELECT  
DISTINCT od."ProductID", 
p."UnitPrice", e."EmployeeID"
FROM employeeterritories e
INNER JOIN territories t ON 
e."TerritoryID" = t."TerritoryID"
INNER JOIN region2 r ON 
t."RegionID" = r."RegionID"
INNER JOIN orders o ON
e."EmployeeID" = o."EmployeeID"
INNER JOIN order_details od ON
o."OrderID" = od."OrderID"
INNER JOIN products p ON 
od."ProductID" = p."ProductID" 
ORDER BY 1, 2) a1
GROUP BY 1
ORDER BY 2;
```

#### 5. What is the total price of units sold per TerritoryDescription (Boston, Denver, etc.)?

```python
%%sql 
SELECT 
DISTINCT od."ProductID", 
p."UnitPrice", 
e."EmployeeID",  
t."TerritoryDescription"
FROM employeeterritories e
INNER JOIN territories t ON 
e."TerritoryID" = t."TerritoryID"
INNER JOIN region2 r ON 
t."RegionID" = r."RegionID"
INNER JOIN orders o ON
e."EmployeeID" = o."EmployeeID"
INNER JOIN order_details od ON
o."OrderID" = od."OrderID"
INNER JOIN products p ON 
od."ProductID" = p."ProductID"
WHERE e."EmployeeID" = 1 
ORDER BY 3, 1;
```

```python
%%sql 
SELECT a1."ProductID", a1."EmployeeID", SUM(a1."UnitPrice") AS "Sum Unit Price" FROM
(SELECT 
DISTINCT od."ProductID", 
p."UnitPrice", 
e."EmployeeID",  
t."TerritoryDescription"
FROM employeeterritories e
INNER JOIN territories t ON 
e."TerritoryID" = t."TerritoryID"
INNER JOIN region2 r ON 
t."RegionID" = r."RegionID"
INNER JOIN orders o ON
e."EmployeeID" = o."EmployeeID"
INNER JOIN order_details od ON
o."OrderID" = od."OrderID"
INNER JOIN products p ON 
od."ProductID" = p."ProductID") a1
GROUP BY 1, 2
ORDER BY 2, 1;
```

```python
%%sql 
SELECT COUNT(DISTINCT a1."TerritoryDescription") AS "Territory count", a1."EmployeeID" FROM
(SELECT 
DISTINCT od."ProductID", 
p."UnitPrice", 
e."EmployeeID",  
t."TerritoryDescription"
FROM employeeterritories e
INNER JOIN territories t ON 
e."TerritoryID" = t."TerritoryID"
INNER JOIN region2 r ON 
t."RegionID" = r."RegionID"
INNER JOIN orders o ON
e."EmployeeID" = o."EmployeeID"
INNER JOIN order_details od ON
o."OrderID" = od."OrderID"
INNER JOIN products p ON 
od."ProductID" = p."ProductID") a1
GROUP BY 2
ORDER BY 2;
```
