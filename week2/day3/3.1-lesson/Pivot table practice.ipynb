```python
import pandas as pd
import numpy as np
df = pd.read_excel("sales-funnel.xlsx")
df.head()
```

```python
df.pivot_table(index="Name",values=["Price","Quantity"])
```

```python
df.dtypes
```

```python
df['Status'] = df['Status'].astype("category")
df['Account'] = df['Account'].astype("category")
```

```python
pd.pivot_table(df,index=["Rep"]) # Default aggfunc is mean
```

```python
pd.pivot_table(df,index=["Manager", "Rep", "Name"],values=["Price"],aggfunc=sum)
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=[np.sum,len])
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=['Product'],aggfunc=[len])
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=['Product'],aggfunc=[np.sum],fill_value=0)
```

```python
pd.pivot_table(df,index=['Manager','Rep','Product'],
              values=['Price'],aggfunc=[np.sum,len],fill_value=0,
              margins=True)
```

```python
pd.pivot_table(df,index=['Status'],columns=['Manager'],
              values=['Price'],aggfunc=[np.sum,len],margins=True)
```

```python
table = pd.pivot_table(df, index=['Manager','Status'],
                      columns=['Product'],values=['Quantity','Price'],
                      aggfunc={'Quantity':len,'Price':[np.sum,len]},
                      fill_value=0)
```

```python
table.query('Manager == ["Debra Henley"] & Status == ["declined"]')
```

```python
table.query('Status != ["won"]')
```

```python
df2.Channel = df2.Channel.astype("category")
df2.Region = df2.Region.astype("category")
```

```python
pd.pivot_table(df2,index=["Region"],columns=["Channel"],aggfunc=np.sum,margins=True)
```

```python
df2['Total'] = df2.ix[:,2:8].apply(np.sum,axis=1)
```

```python
pd.pivot_table(df2,index=["Region"],columns=["Channel"],values=["Total"],aggfunc=np.sum,margins=True)
```

```python
tab = pd.pivot_table(df2,index=["Region"],columns=["Channel"],aggfunc=np.sum)
```
